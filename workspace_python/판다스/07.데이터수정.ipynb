```python
import pandas as pd

df = pd.read_excel('test1.xlsx')
df
```

```python
# 모든 학생의 영어줌수를 현재점수 + 5
df['영어점수'] = df['영어점수'] + 5
df
```

```python
# 평균 점수 수정 
df['평균'] = (df['국어점수'] + df['영어점수'])//2
df
```

```python
# 컬럼 추가하기
df['평평평'] = 'fail'
# 컬럼명 변경
df.rename(columns={'평평평' : '결과'}, inplace=True)
df
```

```python
# 특정 조건에 맞는 데이터만 수정.
# 국어, 영어의 합이 170 이상
df.loc[(df['국어점수'] + df['영어점수'])>=170, '결과'] = 'Pass'
df
```

```python
# 컬럼 삭제
df.drop(columns={'결과'}, inplace=True)
df
```

```python
# 학번을 '번'이라는 글자 제거
# 힌트 : 문자열은 리스트다. int()안에 문자열을 넣으면 숫자로 나온다.

df['학번'] = df['학번'].str.replace('번', '')
df
```
